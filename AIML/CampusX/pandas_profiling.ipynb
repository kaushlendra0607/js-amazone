```python
from google.colab import files
import pandas as pd
import io

# 1. Trigger the file upload widget
uploaded = files.upload()

# 2. Iterate through the uploaded files and load them
datasets = {} # Dictionary to store your DataFrames

for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')

    # Read the file into a Pandas DataFrame
    datasets[filename] = pd.read_csv(io.BytesIO(uploaded[filename]))

# 3. Access a specific dataset by its filename
# Example: If you uploaded 'sales.csv', access it like this:
# df = datasets['sales.csv']
# df.head()
```
